```python
#Data from Table 1:
xs = np.array([201, 244, 47, 287, 203, 58, 210, 202, 198, 158, 165, 201, 157, 131, 166, 160, 186, 125, 218, 146])
ys = np.array([592, 401, 583, 402, 495, 173, 479, 504, 510, 416, 393, 442, 317, 311, 400, 337, 423, 334, 533, 344])
sigma_ys = np.array([61, 25, 38, 15, 21, 15, 27, 14, 30, 16, 14, 25, 52, 16, 34, 31, 42, 26, 16, 22])
```

How do we fit data to a mean model? First, write down the log likelihood function:

$\mathcal{L}(m,b) = \sum_{i=1}^{\infty} \ln p(y_i|r_1,r_2, ...;x_i, \sigma_i)$ where $\{r_1, r_2, ...\}$are parameters of the mean model we are fitting to. For a straight line, we are fitting to the mean model $y = mx+b$, so these parameters are $\{m, b\}$. 

Then write down the probability of observing data point $y_i$ given the mean model and variance of data point $y_i$:

$p(y_i|...) = \mathcal{N}(y_i|mx_i+b, \sigma_i^2) = \frac{1}{\sqrt{2\pi\sigma_i^2}}e^{-\frac{1}{2\sigma_i^2}\left(y_i-[mx_i+b]\right)^2}$

Then the log of this probability is $\ln p(y_i|...) = -\frac{1}{2\sigma_i^2}(y_i-(mx_i+b))^2 + \ln ...$. We don't care about the constant term because we are going to optimize the log likelihood.

Then the log likelihood is $\mathcal{L} = -\frac{1}{2}\sum_{i=1}^{N}\frac{1}{\sigma_i^2}\left(y_i-[mx_i+b]\right)^2+constant$.

Putting this in matrix form, we get $\mathcal{L} = -\frac{1}{2}(Y-X\theta)^TC^{-1}(Y-X\theta)$.

Maximizing the log likelihood, we get $\hat{\theta} = (X^TC^{-1}X)^{-1}X^TC^{-1}Y$.

Realistically, we want to avoid taking inverses in our code, so we'll solve this form using scipy.linalg.solve:

$(X^TC^{-1}X)\hat{\theta} = X^TC^{-1}Y$

noting that scipy.linalg.solve solves the equation $AZ = B$ for $Z$. Also recall that for the diagonal matrix $C$, $C^{-1}$ is equal to taking the inverse of every element along the diagonal. Remember that diagonal matrices work like numbers in terms of operation rules!

The variance on $\hat{\theta}$ is given by the covariance matrix matrix $C_{\hat{\theta}} = (X^TC^{-1}X)^{-1}$. The diagonal elements are the variance of the b parameter and the slope parameter, from top to bottom respectively. 

```python
# define Z, A, and B to plug into scipy.linalg.solve
#Z = theta_hat

#select data
data_xs = xs[4:]
data_ys = ys[4:]
data_sigma_ys = sigma_ys[4:]

def plot_ellipses(data_xs, data_ys, data_sigma_ys, label):
    # first, do the linear fit 
    X = np.vstack((data_xs, np.ones(len(data_xs)))).T
    Y = np.expand_dims(data_ys, 1)
    C_inv = np.zeros((len(data_xs), len(data_xs)))
    np.fill_diagonal(C_inv, 1/data_sigma_ys**2)
    A = (X.T@C_inv)@X
    B = (X.T@C_inv)@Y

    #solve
    Z = linalg.solve(A, B)

    #fit variance
    A = (X.T@C_inv)@X
    B = np.identity(A.shape[0])
    Z_var = linalg.solve(A, B)
    m_var = Z_var[1,1]

    #unpack parameters:
    b = Z[0][0]
    b_err = np.sqrt(Z_var[0][0])  
    m = Z[1][0]
    m_err = np.sqrt(Z_var[1][1])

    # this is the covariance matrix
    C = np.linalg.inv((X.T@C_inv)@X)
    a = C[0][0]
    b = C[1][0]
    c = C[1][1]

    # lambda_1 and lambda_2 are the eigenvalues of the covariance matrix
    # the ellipse is formed semimajor and semiminor axes lambda_1 and lambda_2
    # the ellipse is tilted at the angle theta
    # see https://cookierobotics.com/007/
    lambda_1 = (a+c)/2+np.sqrt(((a-c)/2)**2+b**2)
    lambda_2= (a+c)/2-np.sqrt(((a-c)/2)**2+b**2)
    theta = np.arctan2(lambda_1-a, b)

    r1 = np.sqrt(lambda_2)
    r2 = np.sqrt(lambda_1)
    T = np.linspace(0, 2*np.pi, 1000)

    x_ellipse = [r1*np.cos(theta)*np.cos(t)-r2*np.sin(theta)*np.sin(t) for t in T]
    y_ellipse = [r1*np.sin(theta)*np.cos(t)+r2*np.cos(theta)*np.sin(t) for t in T]

    plt.scatter(m,b, zorder=0,label = label)
    plt.plot(x_ellipse+m,y_ellipse+b)
    plt.xlabel('m', fontsize = 14)
    plt.ylabel('b', fontsize = 14)
```

```python
plot_ellipses(data_xs, data_ys, data_sigma_ys, 'no outliers')
plot_ellipses(xs, ys, sigma_ys, 'outliers')
plt.legend()
```

Comments on the above ellipse plot: Note that the confidence ellipse for the outliers dataset is larger than that of the no outliers dataset. This is because our mean model assumes Gaussian errors, and the outliers dataset violates this assumption.

### Problem 2: Go back to problem 2 of Problem Set 2. What is the standard uncertainty on the period T of the second signal in data set 11 (the last of the 11 data sets)? Get this standard error by an analysis of the profile likelihood, profiling out all other parameters (including all amplitudes).


#### Problem 2 from Problem Set 2: Which of the data sets contain another statistically significant sinusoidal signal at another period? What is the amplitude and period of that signal in each significant case? When you make those measurements, do so in a way that provides uncertainties, marginalizing out the 100-day-period signals.


##### Solution: This profiling method is useful when we have parameters we want to fit in nonlinear terms. It allows us to use the reliable linear fitting method! If there are 2 or more parameters we need to profile, then we need to find the minimum likelihood in 2 or more dimensions (i.e. instead of making $\mathcal{L}$ vs $\theta_1$ plot, make $\mathcal{L}$ vs $\theta_1$ vs $\theta_2...$ plot)

Here our model is a signal consisting of a linear superposition of the 100-day period signal and some other signal of unknown period. 

$y(x) = A_1\cos(\omega_1 x)+A_2\sin(\omega_1 x)+ B_1\cos(\omega_2 x)+B_2\sin(\omega_2 x)$

where $\omega_1 = 2\pi/100$.

We will proceed via brute force: choose a range of possible frequencies $\omega_2$ for the second signal. For each frequency value, do the original, linear fit and then save the log likelihood function using the optimal parameters found. Then plot the log likelihood vs period plot and find the minimum to find the period!

The uncertainty here can be calculated by calculating $-\mathcal{L}(\theta)-(-\mathcal{L}(\hat{\theta})) âˆ¼ \frac{\chi^2}{2}$ where $\hat{\theta}$ are the optimal parameters and $\theta$ are choices of the parameter of interest. We want to find the parameter range $\theta \in [a, b]$ that results in $-\mathcal{L}(\theta)-(-\mathcal{L}(\hat{\theta}))$ less than $\frac{\chi^2(\alpha)}{2}$ where $\alpha = 0.95$ for a 95% confidence interval. That is, we are 95% sure the true value of the desired parameter $\theta$ lies within this parameter range $\theta \in [a, b]$. 

```python
#load in the data
with open("data.pkl", "rb") as handle:
    data = pkl.load(handle).T # check out that pesky little `.T`
print(data.shape)

xs = data[0, :]
yerrs = data[1, :]
ys = data[12, :] - np.mean(data[12, :])
```

```python
def model2_fit(data_xs, data_ys, data_sigma_ys, omegas):
    """does fit to model described in text above, returns parameters, their errors, and negative log likelihood""" 
    omega1, omega2 = omegas
    X = np.vstack((np.cos(omega1*data_xs), np.sin(omega1*data_xs), np.cos(omega2*data_xs), np.sin(omega2*data_xs))).T
    Y = np.expand_dims(data_ys, 1)
    C_inv = np.zeros((len(data_xs), len(data_xs)))
    np.fill_diagonal(C_inv, 1/data_sigma_ys**2)
    A = (X.T@C_inv)@X
    B = (X.T@C_inv)@Y

    #solve
    Z = linalg.solve(A, B)

    #fit variance
    A = (X.T@C_inv)@X
    B = np.identity(A.shape[0])
    Z_var = linalg.solve(A, B)
    
    #$\mathcal{L} = -\frac{1}{2}(Y-X\theta)^TC^{-1}(Y-X\theta)$
    ll = -0.5*(Y-X@Z).T@C_inv@(Y-X@Z)
    
    
    #unpack data
    A1 = Z[0][0]
    A1_err = np.sqrt(Z_var[0][0])  
    A2 = Z[1][0]
    A2_err = np.sqrt(Z_var[1][1])
    B1 = Z[2][0]
    B1_err = np.sqrt(Z_var[2][2])  
    B2 = Z[3][0]
    B2_err = np.sqrt(Z_var[3][3])
    #print(f'The uncertainty variance of the amplitude is {C1_err**2}')
    return A1, A1_err, A2, A2_err, B1, B1_err, B2, B2_err, -ll[0][0]
```

```python
def model2(x, A1, A2, B1, B2, omega1, omega2):
    func = A1*np.cos(omega1*x)+A2*np.sin(omega1*x)+B1*np.cos(omega2*x)+\
            B2*np.sin(omega2*x)
    return func
```

```python
def neg_log_likelihood(data_xs, data_ys, data_sigma_ys, omegas):
    """calculates NLL for model 2 with input data"""
    omega1, omega2 = omegas
    X = np.vstack((np.cos(omega1*data_xs), np.sin(omega1*data_xs), np.cos(omega2*data_xs), np.sin(omega2*data_xs))).T
    Y = np.expand_dims(data_ys, 1)
    C_inv = np.zeros((len(data_xs), len(data_xs)))
    np.fill_diagonal(C_inv, 1/data_sigma_ys**2)
    A = (X.T@C_inv)@X
    B = (X.T@C_inv)@Y
    #solve
    Z = linalg.solve(A, B)
    #$\mathcal{L} = -\frac{1}{2}(Y-X\theta)^TC^{-1}(Y-X\theta)$
    ll = -0.5*(Y-X@Z).T@C_inv@(Y-X@Z)
    return -ll[0][0]
```

```python
#do the profiling
omega_1 = 2*np.pi/100
omega_2s = np.linspace(0.01, 0.2, 4000)
nlls = []
A1s = []
A2s = []
B1s = []
B2s = []

for omega_2 in omega_2s:
    omegas = [omega_1, omega_2]
    A1, A1_err, A2, A2_err, B1, B1_err, B2, B2_err, nll = model2_fit(xs, ys, yerrs, omegas)
    params = [A1, A2, B1, B2]
    A1s.append(A1)
    A2s.append(A2)
    B1s.append(B1)
    B2s.append(B2)
    nlls.append(nll)
```

```python
#plot chi squared vs desired parameter omega_2
plt.semilogy(omega_2s,nlls)
plt.ylabel(r'$\frac{\chi^2}{2}$', fontsize = 20)
plt.xlabel(r'$\omega_2$', fontsize = 14)
```

```python
# plot fit and data
wh_min = np.where(nlls == np.min(nlls))[0][0]
omega_2 = omega_2s[wh_min]
omegas = [omega_1, omega_2]
params = [A1s[wh_min], A2s[wh_min], B1s[wh_min], B2s[wh_min]]
plt.errorbar(xs, ys, yerr=yerrs, color="deeppink", marker="o", alpha=0.2, zorder = 0)
plt.plot(xs,  model2(xs, *params,*omegas), color = 'black', zorder = 1)
plt.xlabel("time")
plt.ylabel("signal")
```

```python
# calculate the error
nll_min = neg_log_likelihood(xs, ys, yerrs, omegas)
profile = (nlls-nll_min)
plt.plot(omega_2s,profile, color = 'black')
wh_2 = np.where(np.abs(profile-1.92)<1)[0]
plt.vlines(omega_2s[wh_2.tolist()[0]], 0, 4, color = 'green', linestyle = '--')
plt.vlines(omega_2s[wh_2.tolist()[1]], 0, 4, color = 'blue', linestyle = '--')
plt.hlines(1.92, 0.023, 0.026, color = 'red', linestyle = '--')
plt.ylim(0,4)
plt.xlim(0.023,0.026)
plt.ylabel(r'$\frac{\chi^2}{2}$', fontsize = 20)
plt.xlabel(r'$\omega_2$', fontsize = 14)
```

```python
#print error: 95% CI
left = omega_2s[wh_2.tolist()[0]]
right = omega_2s[wh_2.tolist()[1]]
print(f'omega_2 = {np.around(omega_2,4)} +/- {np.around(right-left,4)} rad/day')
```

##### M-H MCMC

We want to sample the function $f(\theta) = p(\mathcal{D}|\theta)p(\theta)$.

We know the likelihood $p(\mathcal{D}|\theta)$.

From Wikipedia (https://en.wikipedia.org/wiki/Metropolis%E2%80%93Hastings_algorithm):

Initialization: Choose an arbitrary point $\theta_{i}$ to be the first observation in the sample and choose an arbitrary probability density $q(\theta' \mid \theta)$ that suggests a candidate for the next sample value $\theta'$, given the previous sample value $\theta$. In this section, $q$ is assumed to be symmetric; in other words, it must satisfy $\displaystyle q(\theta'\mid \theta)=q(\theta\mid \theta')$. A usual choice is to let $q$ be a Gaussian distribution centered at $\theta$, so that points closer to $\theta$ are more likely to be visited next, making the sequence of samples into a random walk. The function $q$ is referred to as the proposal density or jumping distribution.

We have generated $k$ samples $\{\theta_{i=1}^{k}\}$ To generate the next sample $\theta_{k+1}$ do the following:
1) Draw a proposal $\theta '$ from the proposal pdf $q(\theta '|\theta_k)$

2) Draw a random number $0 < r < 1$ from the uniform distribution.

3) If $f(\theta')/f(\theta_k) > r$ then $\theta_{k+1}\leftarrow \theta'$; otherwise $\theta_{k+1}\leftarrow \theta_k$.

```python
def f(theta_prime, mean = 2, var = 2):
    """target distribution"""
    return (1/(np.sqrt(var*2*np.pi)))*np.exp(-(0.5*(theta_prime-mean)**2)/var)
```

```python
# run sampler
N = 10**4
theta = 0
thetas = [theta]
while len(thetas)<N:
    theta_prime = np.random.normal(loc=theta, scale=1.0)
    r = np.random.uniform()
    g = np.min((f(theta_prime, mean = 2, var = 2)/f(theta, mean = 2, var = 2),1))
    if g>r:
        theta = theta_prime
    thetas.append(theta)
```

```python
plt.hist(thetas, bins = 50, density = True, alpha = 0.5);
x = np.linspace(-10, 10, 100)
plt.plot(x, f(x, mean = 2, var = 2), color = 'black')
plt.xlabel('theta')
plt.ylabel('p(theta)')
```

#### Problem 4: Re-do Problem 2 but now with an input density that is a function of two variables (x, y). For the density function use two different functions. (a) The first density function is a covariant two-dimensional Gaussian density with variance tensor  [[2.0, 1.2], [1.2, 2.0].

```python
def f_multi(theta_prime): 
    """draw from target function"""
    rv   = multivariate_normal([2,2],  [[2.0, 1.2], [1.2, 2.0]])
    Z    = rv.pdf(theta_prime)
    return Z
```

```python
# do the multivariate MCMC
N = 10**4
theta1 = 0
theta2 = 0
theta1s = [theta1]
theta2s = [theta2]
times = [0]
i = 0
while len(theta1s)<N:
    thetas = [theta1, theta2]
    q   = multivariate_normal(thetas,  [[1,0], [0,1]])
    theta_prime    = q.rvs()
    r = np.random.uniform() # make this 2d
    g= np.min(((f_multi(theta_prime)/f_multi(thetas)),1))
    if g>r:
        theta1, theta2 = theta_prime
    theta1s.append(theta1)
    theta2s.append(theta2)
    i += 1
    times.append(i)
    
```

```python
# plot the sampled contours (filled) and true contours (lines)
X, Y = np.meshgrid(np.linspace(-6,6, 1000), np.linspace(-6,6, 1000))
pos  = np.dstack((X, Y))
rv   = multivariate_normal([2, 2],  [[2.0, 1.2], [1.2, 2.0]])
Z    = rv.pdf(pos)
plt.contour(X, Y, Z, cmap = 'jet')
counts,ybins,xbins,image = plt.hist2d(theta1s,theta2s,bins=20,norm=LogNorm())
plt.contourf(counts,extent=[xbins.min(),xbins.max(),ybins.min(),ybins.max()],linewidths=3)
plt.xlabel('theta1')
plt.ylabel('theta1')
plt.title('p(theta1, theta2)')
```
